# Why Numpy? Why not Python lists?







let us create a numpy array and a normal python list 


*   To create a python list we will simply define a **for loop** from range 0 to 100 and append all the elements into the list.
*   To create a numpy array, firstly we will import the numpy package and will use the arange() function which will basically create a numpy array of size 100
(0 to 100 by default) 




```python
import numpy as np
python_list = [i for i in range(0,100)]
numpy_array = np.arange(100)
print(python_list)
print(numpy_array)
```

**Now let us check the memory requirement of both the arrays.**

**For Numpy Array**

*    arr.itemsize will give size of one element in the numpy array
*   To find size of the whole numpy array we will simply multiply the size of one element in the numpy array with the length of the array



```python
print("size of one element in the numpy array is", numpy_array.itemsize)  
print("the size of the numpy array is",numpy_array.itemsize*numpy_array.size)
```

**For Python List**


*   we will import the package 'sys' and use the function getsizeof() to find the size of python list


```python
import sys
print("the size of the python list is",sys.getsizeof(1)*len(python_list))
```

Thus we can see that the memory consumed by numpy array is 800 whereas the memory consumed by python list is 2800. Thus memory managment of Numpy is way better than python list that's why we prefer numpy for computation of large amount of data.

**We can also compare the Time complexity of both numpy array and python list**

```python
import time
import numpy as np
size = 10000

def add_using_list():
  t1 = time.time()
  a = range(size)
  b = range(size)
  add = [a[i]+b[i] for i in range(size)]
  t2 = time.time()
  return (t2-t1)*1000

def add_using_numpy():
  t1=time.time()
  a = np.arange(size)
  b = np.arange(size)
  c = a+b
  t2 = time.time()
  return (t2-t1)*1000
  
print("time taken by list in ms",add_using_list())
print("time taken by numpy array in ms",add_using_numpy())
```

Hence we can see that TC of **numpy < python list** which makes it a better way to compute large amount of numerical data.

# np.zeros() and np.empty()
NumPy arrays cannot grow the way a Python list does: No space is reserved at the end of the array to facilitate quick appends. So it is a common practice to either grow a Python list and convert it to a NumPy array when it is ready or to preallocate the necessary space with np.zeros or np.empty


*   np.zeros((shape),dtype='int' or 'float')
*   np.empty((shape),dtype='int' or 'float')

**NOTE** if you dont specify the dtype in empty() all the values will be taken as zero by default

# Numpy Arithmetic Operations
Arithmetic is one of the places where NumPy speed shines most. Vector operators are shifted to the c++ level and allow us to avoid the costs of slow Python loops. NumPy allows the manipulation of whole arrays just like ordinary numbers

6.png

```python
import numpy as np
a = np.array([1,2,3])
b = np.array([3,4,5])

sum = a+b
print("sum =",sum)
subtraction = a-b
print("subtraction = ",subtraction)
dv = a/b       #will give output in float64 dtype
print("division = ",dv) 
dv2 = a//b #will give output in int32 dtype
print("int division =",dv2)
```

Numpy is also fully able to perform math operations like square root, exponent, log etc and have pre-defined functions for it. visit the documentation to see those functions as they are pretty straight forward.

# Axis
In many operations (e.g., sum) you need to tell NumPy if you want to operate across rows or columns you specify axis for this.
the first dimension i (axis=0) is responsible for indexing the rows and for   the first dimension i (axis=1) is responsible for indexing the columns.

7.png

```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
s = arr.sum(axis=0)
s2 = arr.sum(axis=1)
print(s,s2)
```

# boolean indexing


```python
import numpy as np
b = np.random.randint(1,20,6)
print(b)  #prints the random array b
print(b>10)   #prints the boolean array with the condition b>10 as true
print(b[b>10])   #prints the elements of the array b satisfying the condition b>10
print(b[(b>10) & (b<17)])  #prints the elements of the array b satisfying the condition b>10 and b<17
```
